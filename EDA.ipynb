```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the countplot
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")  # Set background style for a cleaner look

# Customizing the count plot
ax = sns.countplot(x='diagnosis', data=br, palette='muted')

# Add title and labels
ax.set_title('Count of Diagnosis Categories', fontsize=16, weight='bold')
ax.set_xlabel('Diagnosis', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

# Add counts on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12,
                color='black', fontweight='bold', xytext=(0, 10),
                textcoords='offset points')

# Adjusting layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 12))

features = [
    ('radius_mean', 'red'),
    ('texture_mean', 'blue'),
    ('perimeter_mean', 'gray'),
    ('area_mean', 'green'),
    ('compactness_mean', 'lightblue'),
    ('concavity_mean', 'salmon'),
    ('symmetry_mean', 'lightgreen'),
    ('fractal_dimension_mean', 'orange')
]

for i, (feature, color) in enumerate(features):
    plt.subplot(4, 2, i + 1)
    sns.boxplot(x=br[feature], color=color)
    plt.title(f"Boxplot - {feature}")

plt.tight_layout()
plt.show()

```

   <b>Handling outliers using IQR</b>
</div>


```python
def check_outliers(col,br):
    q1 = br[col].quantile(0.25)
    q3 = br[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    outlier_in_df = br[(br[col] < lower_bound) | (br[col] > upper_bound)].shape[0]
    if outlier_in_df > 0:
        print(f'{col} contains outliers with count : {outlier_in_df}')
    else:
        print(f"{col} Does'nt contain outliers")    
```

```python
def remove_outliers(col, br):
    q1 = br[col].quantile(0.25)
    q3 = br[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    outlier_mask = (br[col] < lower_bound) | (br[col] > upper_bound)
    outlier_count = outlier_mask.sum()

    if outlier_count > 0:        
        br = br[~outlier_mask]  # Keep only rows that are NOT outliers
        print(f"Outliers removed from {col}")
    else:
        print(f"{col} doesn’t contain outliers")

    return br
```

```python
for col in br.select_dtypes(['int','float']).drop('id',axis = 1):
    check_outliers(col,br)
```

```python
for col in br.select_dtypes(['int','float']).drop('id',axis = 1):
    br = remove_outliers(col,br)
```
